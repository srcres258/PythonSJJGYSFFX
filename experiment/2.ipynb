实验内容1：分形树

```python
import turtle as tt

BASE_LEN = 100
MAX_LEVEL = 7

def draw_tree(level: int) -> None:
    tt.fd(BASE_LEN / level)
    if level < MAX_LEVEL:
        tt.lt(60)
        draw_tree(level + 1)
        tt.rt(120)
        draw_tree(level + 1)
        tt.lt(60)
    tt.back(BASE_LEN / level)
```

```python
tt.clearscreen()
tt.lt(90)
draw_tree(1)
```

实验内容2：格雷码构造问题

```python
from tqdm.notebook import tqdm

def normal_to_gray(normal: int) -> int:
    """普通码转格雷码"""
    
    return normal ^ (normal >> 1)

def gray_to_normal(gray: int) -> int:
    """格雷码转普通码"""
    
    return gray ^ (gray >> 1)
```

```python
for i in tqdm(range(256), desc='Gray Code'):
    gray = normal_to_gray(i)
    print(f"Normal: {i} ({bin(i)}), Gray: {gray} ({bin(gray)})")
```
