## 实验项目名称

算法时间复杂度渐进分析

## 目的、要求

通过 `time` 模块，验证 Python 的 `list` 和 `dict` 各种操作的时间复杂度。掌握时间复杂度的渐进分析。用 `matplotlib` 模块画图。

## 实验内容

做计时实验，验证时间复杂度：

1. 验证 `list` 的按需索引：$O(1)$。

2. `dict` 的 set item 和 get item：$O(1)$。

3. 给定随机数列表，`list.sort()`：$O(n \log n)$。


实验内容1：验证 `list` 的按需索引：$O(1)$。

```python
from matplotlib import pyplot as plt
from typing import List
import random
import time

def gen_random_list(n: int) -> List[int]:
    """生成给定大小规模，内容随机的列表。"""
    
    return [random.randint(0, 10000) for i in range(n)]
```

```python
print("Generating random list...")
my_list = gen_random_list(2000000)
x = []
y = []
print("Calculating time taken for accessing each element in the list...")
for i in range(len(my_list)):
    start = time.time()
    val = my_list[i]
    end = time.time()
    x.append(i)
    y.append(end - start)
print("Plotting out the data...")
plt.figure()
plt.plot(x, y)
plt.show()
```

实验内容2：`dict` 的 set item 和 get item：$O(1)$。

```python
from matplotlib import pyplot as plt
import random
import time

my_dict = dict()
print("Setting up dict and calculating time cost...")
set_x = []
set_y = []
for i in range(1000000):
    start = time.time()
    my_dict[i] = random.randint(1, 10000)
    end = time.time()
    set_x.append(i)
    set_y.append(end - start)
print("Accessing dict and calculating time cost...")
get_x = []
get_y = []
for i in range(len(my_dict)):
    start = time.time()
    val = my_dict[i]
    end = time.time()
    get_x.append(i)
    get_y.append(end - start)
print("Plotting time cost...")
plt.figure()
plt.plot(set_x, set_y)
plt.title("set item")
plt.figure()
plt.plot(get_x, get_y)
plt.title("get item")
plt.show()
```

实验内容3：给定随机数列表，`list.sort()`：$O(n \log n)$。

```python
from matplotlib import pyplot as plt
from typing import List
import random
import time
from tqdm.notebook import tqdm

def gen_random_list(n: int) -> List[int]:
    """生成给定大小规模，内容随机的列表。"""
    
    return [random.randint(0, 10000) for i in range(n)]
```

```python
print("Start generating lists and calculating the sorting time cost for each list...")
x = []
y = []
for i in tqdm(range(1000000, 10000001, 1000000), desc="Processing"):
    my_list = gen_random_list(i)
    start_time = time.time()
    my_list.sort()
    end_time = time.time()
    x.append(i)
    y.append(end_time - start_time)
print("Plotting the sorting time cost for each list...")
plt.figure()
plt.plot(x, y)
plt.show()
```
