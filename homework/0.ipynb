1. 设计一个实验，证明：列表的索引操作为常数阶的时间复杂度。

```python
from matplotlib import pyplot as plt
import numpy as np
import time

data = np.random.rand(500000)

def run(scale: int) -> float:
    """
    给定数据规模，执行列表的索引操作并返回所用时间。
    """
    
    index = np.random.randint(0, scale)
    start_time = time.time()
    item = data[index]
    end_time = time.time()
    print(f"{item}, ", end="")
    return end_time - start_time

if __name__ == "__main__":
    x = []
    y = []
    for scale in range(20, 500000, 20):
        result = run(scale)
        x.append(scale)
        y.append(result)
    print()
    npx = np.array(x)
    npy = np.array(y)
    plt.figure()
    plt.plot(npx, npy)
    plt.show()
```

2. 设计一个实验，证明字典的取操作和赋值操作为常数阶时间复杂度。

```python
from matplotlib import pyplot as plt
import numpy as np
import time

data = { i: str(i) for i in range(500000) }

def run_dict_get_value(scale: int) -> float:
    """
    给定数据规模，执行字典的取操作并返回所用时间。
    """
    
    index = np.random.randint(0, scale)
    start_time = time.time()
    item = data[index]
    end_time = time.time()
    print(f"{item}, ", end='')
    return end_time - start_time

def run_dict_set_value(scale: int) -> float:
    """
    给定数据规模，执行字典的赋值操作并返回所用时间。
    """
    
    index = np.random.randint(0, scale)
    start_time = time.time()
    data[index] = "114514"
    end_time = time.time()
    return end_time - start_time

if __name__ == "__main__":
    x1 = []
    y1 = []
    x2 = []
    y2 = []
    for scale in range(20, 500000, 20):
        result = run_dict_get_value(scale)
        x1.append(scale)
        y1.append(result)
        result = run_dict_set_value(scale)
        x2.append(scale)
        y2.append(result)
    print()
    npx1 = np.array(x1)
    npy1 = np.array(y1)
    npx2 = np.array(x2)
    npy2 = np.array(y2)
    plt.figure()
    plt.plot(npx1, npy1, label="get")
    plt.figure()
    plt.plot(npx2, npy2, label="set")
    plt.show()
```

3. 给定一个随机排列的数字列表，编写一个找出第k小的元素的O(nlogn)算法。

答：使用 Timsort 排序算法，时间复杂度为 O(nlogn)。该算法实质是插入排序与归并排序的综合。

```python
from typing import List

def insertion_sort(arr: List[int], left: int, right: int) -> None:
    """插入排序"""
    
    for i in range(left + 1, right + 1):
        j = i
        while j > left and arr[j] < arr[j - 1]:
            arr[j], arr[j - 1] = arr[j - 1], arr[j]
            j -= 1

def merge_sort(arr: List[int], l: int, m: int, r: int) -> None:
    """归并排序"""
    
    # 将原数组分为左数组和右数组
    len1, len2 = m - l + 1, r - m
    left, right = [], []
    for i in range(0, len1):
        left.append(arr[l + i])
    for i in range(0, len2):
        right.append(arr[m + 1 + i])
        
    i, j, k = 0, 0, l
        
    # 比较后将两个数组合并成一个更大的数组
    while i < len1 and j < len2:
        if left[i] <= right[j]:
            arr[k] = left[i]
            i += 1
        else:
            arr[k] = right[j]
            j += 1
        k += 1
    
    # 复制左数组的剩余元素
    while i < len1:
        arr[k] = left[i]
        i += 1
        k += 1
    
    # 复制右数组的剩余元素
    while j < len2:
        arr[k] = right[j]
        j += 1
        k += 1

MIN_MERGE: int = 32

def calc_min_run(n: int) -> int:
    """Timsort 算法要用到的，计算run块的最小长度，以确保run块的归并可以高效地进行"""
    
    r = 0
    while n >= MIN_MERGE:
        r |= n & 1
        n >>= 1
    return n + r

def timsort(arr: List[int]) -> None:
    """Timsort 算法"""
    
    n = len(arr)
    min_run = calc_min_run(n)
    
    # 对大小为 run 的单个子数组进行排序
    for start in range(0, n, min_run):
        end = min(start + min_run - 1, n - 1)
        insertion_sort(arr, start, end)
    
    # 从大小 run （或32）开始合并。最终合并形成大小为 2^n
    size = min_run
    while size < n:
        # 喧杂左子数组的起点。
        # 合并 arr[left .. left + size - 1] 和 arr[left + size .. left + 2 * size - 1]
        for left in range(0, n, 2 * size):
            # 查找左子数组的终点
            # mid + 1 为右子数组的起点
            mid = min(n - 1, left + size - 1)
            right = min((left + 2 * size - 1), (n - 1))
            # 合并 arr[left .. mid] 和 arr[mid + 1 .. right]
            if mid < right:
                merge_sort(arr, left, mid, right)
        size *= 2

if __name__ == '__main__':
    arr = [-2, 7, 15, -14, 0, 15, 0, 7, -7, -4, -13, 5, 8, -14, 12]
    
    print("排序前：")
    print(arr)
    
    timsort(arr)
    
    print("排序后：")
    print(arr)
```

```python
from matplotlib import pyplot as plt
import numpy as np
import time

data = [500000 - i for i in range(500000)]

def run_timsort(scale: int) -> float:
    """给定数据规模，运行 Timsort 排序算法并返回运行时间。"""
    
    arr = data[:scale].copy()
    start_time = time.time()
    timsort(arr)
    end_time = time.time()
    return end_time - start_time

if __name__ == '__main__':
    x = []
    y = []
    for scale in range(1000, 100000, 500):
        result = run_timsort(scale)
        x.append(scale)
        y.append(result)
    
    plt.figure()
    plt.plot(x, y)
    plt.show()
```
